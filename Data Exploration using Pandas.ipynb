# 9. Pivot Table: 

```python
#As we build up the pivot table, I think it’s easiest to take it one step at a time. 
#Add items and check each step to verify you are getting the results you expect. Don’t be afraid to play with the order and the variables to see what presentation makes the most sense for your needs.
#It helps to generate data structure. It has three components index, columns and values (similar to excel)
#By default, it shows the sum of values column but you can change it using argument aggfunc

#The simplest pivot table must have a dataframe and an index .
pd.pivot_table(df,index=["encounter_id"])

pd.pivot_table(df, values='number_diagnoses', 
                   index=['admission_type_id','admission_source_id'], 
                   columns=['readmitted']).head(20)

#pd.pivot_table(df, values='encounter_id', index=[‘encounter_id’,’encounter_id’], columns=[‘encounter_id’])
 
```

```python
#it shows count
pd.pivot_table(df, values='number_diagnoses', 
                   index=['admission_type_id','admission_source_id'], 
                   columns=['readmitted'],aggfunc=len).head(25)
```

```python
###Cross Tab: Cross Tab computes the simple cross tabulation of two factors.
pd.crosstab(df.admission_type_id, df.readmitted)
```

# 12.Applying function to element, column or dataframe

```python
##a. Map: It iterates over each element of a series.
#df[‘column1’].map(lambda x: 10+x) #this will add 10 to each element of column1
df['num_lab_procedures'].head(10)
df4 = df['num_lab_procedures'].map(lambda x: 10+x)
df4.head(10)
```

```python
## #this will concatenate “AV“ at the beginning of each element of column2 (column format is string)
df5 = df['examide'].map(lambda x: 'very_'+x) 
df5.head(10)
```

```python
###b. Apply: As the name suggests, applies a function along any axis of the DataFrame.
## it will returns the sum of all the values of column1 and column2.

df4 = df[['num_medications','num_procedures']].apply(sum) 
df4
```

```python
###c. ApplyMap: This helps to apply a function to each element of dataframe.
#it will add 2 to each element of dataframe (all columns of dataframe must be numeric type)

func = lambda x: x+2
df[['num_medications']].applymap(func).head(10)
```
